# The Use of Crushed Concrete as an Aggregate in Concrete



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diameter = 4                                     # in
cross_section = np.pi*diameter**2/4              # in**2

Recyc_force = np.array([48450, 41995])           # lb_f
Recyc_pres = Recyc_force/cross_section           # psi
Recyc_avg_pres = np.average(Recyc_pres)          # psi

Control_force = np.array([39615, 50990, 38990])  # lb_f
Control_pres = Control_force/cross_section       # psi
Control_avg_pres = np.average(Control_pres)      # psi
Control_std_pres = np.std(Control_pres)
```

```python
plt.title("Experimental Compressive Strength of Concrete\nvia Cylinder Stress Test")
plt.ylabel(r"Pressure ($psi$)")
plt.boxplot((Recyc_pres, Control_pres), vert=True, labels=("Recycled Concrete (n = 2)","Control (n = 3)"), meanline=False)
plt.savefig("cylinder_test_box_plot.png");
```
