# I. Basic File I/O

File input and output (I/O) is needed for engineering work. You don't want to be typing and copying lots of numbers manually!

## Some Numpy commands
* ```np.savetxt```
* ```np.loadtxt```
* ```np.column_stack```

### Step 1: Create some data

````python
import numpy as np

x     = np.linspace(0,10,21)
y_exp = x**0.6 + (np.random.rand(21) -0.5)        # ploting data series below
y_1   = x**0.6
y_2   = np.exp(-0.2*x)*np.cos(2*x)*2
````

### Step 2: Create a single 2-D array (matrix) for saving to a file

````python
data = np.column_stack([x,y_exp, y_1, y_2])   # combine column vector arrays into a matrix array
print(data)
````

### Step 3: Save to file

````python
np.savetxt("data.dat", data)

#----- OR THIS

np.savetxt("data.dat", data, fmt='%10.3f', # format the data as floats with 3 digits after decimal
           header="       x,     y_exp,       y_1,       y_2",    # put a header row on your data file to describe each column
           delimiter = ',')                # delimiter is the symbol separating columns

!cat data.dat
````

### Step 4: Load from file

````python
loaded_data = np.loadtxt("data.dat", delimiter = ',')    # delimiter needed if used above
print(loaded_data)
````

## General File IO

If you have large data files and you need more control what gets read in or read out

### 1. Input
#### ```readline``` function

````python
ifile = open("data.dat", "r")   # open a file for *reading*
s = ifile.readline()
print(s)
s = ifile.readline()            # readline function will automatically read the next line in the file
print(s)
ifile.close()                   # close the file when you're done
````

#### ```readlines``` function

````python
ifile = open("data.dat", "r")
lines = ifile.readlines()             # readlines function gets a list of all lines
for line in lines :
    print(line, end='')
````

#### ```split``` function

````python
words = line.split()
print(line)
print(words)
for w in words:
    print(w)
````

### 2. Output


````python
ofile = open("output.dat", "w")         # Open a file for writing: ofile can be any name
ofile.write("Some header\n")            # write some text: note the "\n" makes a new line
ofile.write("some new line.")           # write another line of text (note there is no \n)
ofile.write(" This isnt a new line\n")  # and another
ofile.write("The third line")           # and other
ofile.close()                           # close the file

!cat output.dat
````

### Triple strings preserve white space

````python
ofile = open("triplestring.dat", "w")
s = """Hi there,
How is your day going so far?

    I hope you are good."""

ofile.write(s)
ofile.close()
!cat triplestring.dat
````

## 1. Pyplot interface

````python
x = np.linspace(0,6,80)
y = np.exp(-x)*np.sin(4*x)
````

### Basic plot

````python
plt.plot(x,y);   # use ";" on last plot command 
                  # (if that is the last cell command)
                  # to suppress the undesired output: [<matplotlib.lines.Line2D at 0x1113b8e80>] 
````

### Add formatting

````python
plt.plot(x,y+1, '+-g'  )   # linestyle: k is black, o is symbol, - is solid line
                           # Other colors: b,r,g,c,m; other lines: --, -., : other symbols: .s^v*dpx+<>h

plt.xlabel('The x-axis label')   
plt.ylabel('The y-axis label', fontsize=16)
plt.title('Big fat idea');

#help(plt.plot)
````

### Draw multiple curves on the same plot

````python
plt.plot(x,     y+0.5,        '--', color='black', linewidth=2)
plt.plot(x[1:], 1/x[1:]**0.2, 'x-', color='blue',  markersize=6)

plt.yscale('log')                                              
plt.xlim([0,6])
plt.ylim([0.1,10])
plt.xlabel(r'With Symbols: $\alpha_{s}\beta^{s}$', fontsize=20) # r'text $ latex $' formatting required (including preceding r) if you want to use latex
plt.ylabel('some label', fontsize=20)                           
plt.legend(['curve 1', 'curve2'], fontsize=16, frameon=False, loc='upper left');

plt.savefig("plot2.pdf")
````

### Subplots: like a table of plots

````python
plt.rc("font", size=10)        # Set all font sizes

plt.subplot(2,2,1)             # rows, cols, active;  where active is the plot that is currently active
plt.plot(x,y)                  

plt.subplot(2,2,2)             # add labels, legends to each subplot separately
plt.plot(x,y*x)
plt.xlabel('my x')             # format each plot as previously shown

plt.subplot(2,2,3)
plt.plot(x,np.sqrt(x))

plt.subplot(2,2,4)
plt.plot(x,1/(x+1))

plt.tight_layout()             # keep labels from overlapping
````

### Contour plot of 3-D data: z(x,y)

````python
#--------- Set the X,Y,Z data
Lx = 6
Ly = 6
nx = 100 
ny = 100
x = np.linspace(-Lx/2,Lx/2,nx)
y = np.linspace(-Ly/2,Ly/2,ny)
X,Y = np.meshgrid(x,y)
Z = np.sin(2*X)*np.sin(3*Y)*X/Y*np.exp(np.abs(Y))

#---------- make the plot
plt.subplot(1,2,1)
plt.contour(X,Y,Z,20)          # 20 is the number of contour levels chosen
plt.subplot(1,2,2)
plt.contourf(X,Y,Z,20)         # contourf is a filled contour plot
plt.colorbar();                # add a colorbar for the Z scale   
````

````python
import numpy as np
import matplotlib                     # for saving to pdf
matplotlib.use('PDF')                # for saving to pdf
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
x = np.linspace(0,10,25)
y_exp = x**0.6 + (np.random.rand(25) -0.5)        # ploting data series below
y_1   = x**0.6
y_2   = np.sin(x)*2
y_3   = np.exp(-0.2*x)*np.cos(2*x)*2
````

````python
fig, ax = plt.subplots(1,1, figsize=(7,5))     # create a figure and axes
plt.rcParams.update({'font.size': 18})         # change all font sizes (specialize below)

ax.tick_params(direction='in', top=True, right=True)

ax.plot(x, y_1,   '-',  color='gray',  linewidth=4)
ax.plot(x, y_2,   '--', color='blue',  linewidth=1)
ax.plot(x, y_3,   ':',  color='red',   linewidth=1)
ax.plot(x, y_exp, 'o',  color='black', markersize=6);

ax.set_xlabel(r"$v_{rms} + \theta$", fontsize=20)
ax.set_ylabel("position",            fontsize=20)
ax.set_xlim([0,10])
ax.set_ylim([-2,6])
ax.legend(["T (K)", 
           r"$\Delta_{max}=2$", 
           r"$\rho$ (kg/m$^3$)", 
           "Exp"], loc='upper left', frameon=False, fontsize=16);

plt.savefig("plot.pdf")                         # for saving file plot.pdf
````
